```python
#importing librieris
import numpy as np
import pandas as pd
```

```python
#loading data
dataset =pd.read_csv('houses.csv')
#showing data
dataset.head
```

```python
#clean data
from sklearn.impute import SimpleImputer 
imp = SimpleImputer(missing_values = np.nan, strategy='mean')
imp = imp.fit(dataset)
dataset = imp.transform(dataset)

#slicing data
x = dataset[:, :-1]
print('X :\n', x)
print('X shape :',x.shape)
y = dataset[:, -1]
print('y :\n', y)
print('Y shape :',y.shape)
```

```python
#Scaling data to set all data at same range 'rearange data'
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x =sc.fit_transform(x)#only rearangeng x
print('X after rearangeng :\n',x)
```

```python
#spliting data into train and test
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(x ,y ,test_size=0.33,random_state=42 ,shuffle=True)
print('X_train is  ' , x_train)
print('X_train Shape is  ' , x_train.shape)
print('X_test is  ' , x_test)
print('X_test Shape is  ' , x_test.shape)
print('y_train is  ' ,y_train)
print('y_train Shape is  ' ,y_train.shape)
print('y_test is  ' , y_test)
print('y_test Shape is  ' , y_test.shape)
```

```python
# Fitting Polynomial Regression to the dataset
#increasing features for better accurancy
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2) #degree=2 means get column162....columnk^2
X_train2 = poly_reg.fit_transform(x_train) #fit for applying equations for Polynomial Regression 
X_test2 = poly_reg.fit_transform(x_test)
print('X_train new :\n',X_train2)
print('X_test new :\n',X_test2)

# No Polynomial for y

print('shape of X_train new :\n', X_train2.shape )
print('shape of X_test new :\n', X_test2.shape )
```

```python
#Applying linear regretionn algorithm
from sklearn.linear_model import LinearRegression
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_train2, y_train )#work on newe x 'X_train2 and X_test23'

y_pred2 = lin_reg_2.predict(X_test2) 

#calculating erorr
from sklearn.metrics import mean_absolute_error
print('mean_absolute_error :\n', mean_absolute_error(y_test, y_pred2))

from sklearn.metrics import mean_squared_error
print('mean_squared_error :\n', mean_squared_error(y_test, y_pred2))

from sklearn.metrics import median_absolute_error
print('median_absolute_error :\n', median_absolute_error(y_test, y_pred2))
```
